```python
graph = {
    'S': ['A', 'B'],
    'A': ['C', 'D'],
    'B': ['E'],
    'C': [],
    'D': ['G'],
    'E': [],
    'G': []
}

visited = set()
stack = []

start = 'S'
stack.append(start)

print("\nDFS Traversal:")

while stack:
    node = stack.pop()

    if node not in visited:
        print(node, end=" ")
        visited.add(node)

        # reverse for correct order
        for neighbor in reversed(graph[node]):
            if neighbor not in visited:
                stack.append(neighbor)
```
