```python
graph = {
    'S': ['A', 'B'],
    'A': ['C', 'D'],
    'B': ['E'],
    'C': [],
    'D': ['G'],
    'E': [],
    'G': []
}

from collections import deque

visited = set()
queue = deque()

start = 'S'
queue.append(start)

print("BFS Traversal:")

while queue:
    node = queue.popleft()

    if node not in visited:
        print(node, end=" ")
        visited.add(node)

        for x in graph[node]:
            if x not in visited:
                queue.append(x)
```
