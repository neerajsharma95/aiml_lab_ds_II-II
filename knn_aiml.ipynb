```python
import pandas as pd
import math

# Read CSV
data = pd.read_csv("student_data.csv")

# Convert to list of tuples
dataset = list(zip(data['Hours'], data['Attendance'], data['Result']))

print("Dataset:\n", dataset)

# Input
test_hours = float(input("\nEnter study hours: "))
test_attendance = float(input("Enter attendance: "))

# Distance calculation
distances = []

for x, y, label in dataset:

    d = math.sqrt((test_hours - x)**2 + (test_attendance - y)**2)

    distances.append((d, label))

    print(f"Distance from ({x},{y}) = {round(d,2)}")

# Sort
distances.sort()

# K
K = 3
nearest = distances[:K]

print("\nNearest Neighbors:", nearest)

# Voting
votes = {}

for d, label in nearest:
    votes[label] = votes.get(label, 0) + 1

print("Votes:", votes)

# Prediction
prediction = max(votes, key=votes.get)

print("\nFinal Prediction:", prediction)
```
